## SS.lv datu analīze kategorijai vieglie auto

##### Atlasītie dati ir no ss.lv sludinājumiem par vieglajām automašīnām laika posmā no 11.11.2023 līdz 05.01.2024.

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Datu atlase un apstrāde

```python
#get path for data file
path_parent = os.path.join(os.getcwd(),os.pardir)
path_file = path_parent+'\\data\\ss_car_data.csv'
```

```python
df=pd.read_csv(path_file)
df.head()
```

Aplūkojot datu pirmās rindas redzam, ka dati nav attīrīti, nepieciešama datu attīrīšana no datiem, kurus nevarēsim izmantot priekš analīzes
Pēc datu aplūkošanas redzams, ka nepieciešams no datu kopas izmest tās rindas,kur:
- kāda atribūta vērtība nav norādīta
- 'Price' atribūta vērtība ir 'pērku'
- 'Price' atribūta vērtība satur vārdu 'maiņai', jo sludinājumā norādīta neatbilstošā ražotāja auto
- atribūtā 'Milage' jāpārveido vārds 'tūkst.' uz attiecīgo skaitlisko vērtību
- atribūtā 'Price' jānoņem nost Eur zīme un komats,kas atrodas skaitlī
- jānomaina kolonnu tipi no string tipa uz tiem atbilstošiem tipiem
- jāizfiltrē motoru tipi, jānolasa no motora apraksta un jāievieto atsevišķā kolonnā
- ja motora tips ir E (elektromotors) tad jānomaina 'Motor Liters' vērtība uz 0
- jānoņem motora tipa simbols kas atrodas kolonnā 'Motor Liters'

```python
#data filtering 

filtered_df = df[df['Price'] != 'pērku ']#filters out if Price= 'pērku '
filtered_df = filtered_df[~filtered_df['Price'].str.contains('maiņai')]# filters out if 'maiņai' is in 'Price'

filtered_df = filtered_df[filtered_df['Milage'] != '-']#filters out if 'Milage' is not specified
filtered_df = filtered_df[filtered_df['Model'] != '-']#filters out if 'Model' is not specified
#filtered_df = filtered_df[filtered_df['Manufacturer'] == '-']#filters out if 'Model' is not specified

filtered_df['Milage'] = filtered_df['Milage'].str.replace(' tūkst.','000')#replacing word 'tūkst.' with 000
filtered_df['Price'] = filtered_df['Price'].str.replace(' €','')#removing word ' €' in Price
filtered_df['Price'] = filtered_df['Price'].str.replace(',','')#removing ',' in Price
filtered_df['Milage'] = filtered_df['Milage'].str.replace(',','')#removing ',' in Milage

filtered_df['Price'] = filtered_df['Price'].astype(float)#type change
filtered_df['Milage'] = filtered_df['Milage'].astype(float)#type change
filtered_df['Year'] = filtered_df['Year'].astype(int)#type change

#create new colum for Motor Type and assign correct type
filtered_df = filtered_df.assign(MotorType=filtered_df['Motor Liters'])
#check if type contains letter 'D', if does then change name to Diesel
filtered_df['MotorType'] = np.where(filtered_df['MotorType'].str.contains("D") ,'Diesel',filtered_df['MotorType'])
filtered_df['MotorType'] = np.where(filtered_df['MotorType'].str.contains("E") ,'Electric',filtered_df['MotorType'])
filtered_df['MotorType'] = np.where(filtered_df['MotorType'].str.contains("H") ,'Hybrid',filtered_df['MotorType'])
filtered_df['MotorType'] = np.where(~(filtered_df['MotorType'].str.contains("E") | 
                                      filtered_df['MotorType'].str.contains("D") |
                                      filtered_df['MotorType'].str.contains("H")),
                                    'Gasoline',filtered_df['MotorType'])
#remove symbols H , D and E from Motor Liters column
filtered_df['Motor Liters'] = filtered_df['Motor Liters'].str.replace('D','')#remove symb 'D'
filtered_df['Motor Liters'] = filtered_df['Motor Liters'].str.replace('H','')#remove symb 'H'
filtered_df['Motor Liters'] = filtered_df['Motor Liters'].str.replace('E','0')#replace 'E'with 0

filtered_df['Motor Liters'] = filtered_df['Motor Liters'].astype(float)#type change


filtered_df
```

### Datu analīze

```python
#group by manufacturer and create pie chart
count_manufacturer = filtered_df.groupby('Manufacturer').size()
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot()
plt.pie(count_manufacturer, labels=count_manufacturer.index,autopct='%1.1f%%')
plt.title('Auto ražotāju popularitāte tirgū')
plt.show()
```

```python
#group by motor type and create pie chart
count_motor = filtered_df.groupby('MotorType').size()
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot()
plt.pie(count_motor, labels=count_motor.index,autopct='%1.1f%%')
plt.title('Auto motoru tipu popularitāte tirgū')
plt.show()
```

```python
#group by manufacturing year ranges and create pie chart
bins=[1900,2000,2005,2010,2015,2020,2025]
group_year = filtered_df.groupby(pd.cut(filtered_df['Year'], bins=bins),observed=False)
year_size = group_year['Year'].size()

fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot()
labels = ['līdz 2000','2000-2005','2005-2010','2010-2015','2015-2020','2020-2024']
plt.pie(year_size,labels=labels,autopct='%1.1f%%')
plt.title('Auto reģersrācijas gadi tirgū')
plt.show()
```

```python
#group by year, compare to price, calculate for each motor type separately

car_data = filtered_df.copy()
motor_types=['Diesel','Gasoline','Electric','Hybrid']#motor types

fig = plt.figure(figsize=(9, 5))
ax = fig.add_subplot()

for motor_type in motor_types:#iterate motor types, group by year,get mean price and create scatter plot
    type_data=car_data[car_data['MotorType']==motor_type]
    group_year = type_data.groupby('Year')['Price'].mean()
    year = group_year.index.to_numpy()
    mean = group_year.to_numpy()/1000
    plt.scatter(2024-year,mean, label=motor_type)
#plot atributes
plt.axvline(x=30, color='red', linestyle='dotted', label='Vēsturiskā statusa vecums')
plt.legend(title="Motora tips")
plt.title('Car manufacturing year vs price for different motor types')
plt.title('Auto vidējā cena atkarībā no vecuma \nauto ar dažādiem motora tipiem')
plt.xlabel('Vecums, gadi')
plt.ylabel('Vidējā cena, Eur, tūkst.')
plt.show()
```
